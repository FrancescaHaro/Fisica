# 1. Movimiento de un paracaidista 

#### a)Parámetros: 

```python
import random
resistencia = 1.17
aire = 1.20
masa_parac = 80
area_parac = random.uniform(6.00,10.01)
rapidez_inicial = random.uniform(0.00,5.01)
tiempo_parac = random.uniform(15.00,23.01)
altura_inicial = random.uniform(3000, 3500)
gravedad = 9.81
k = (resistencia * aire * area_parac) / 2

print("Coeficiente de resistencia: ", resistencia)
print("Densidad del aire: ", aire)
print("Masa del paracaidista: ", masa_parac)
print("Area transversal del paracaidista: ", area_parac)
print("Rapidez inicial: ", rapidez_inicial)
print("Tiempo en que se abre el paracaidas: ", tiempo_parac) 
print("Altura inicial: ", altura_inicial)
print("K: ", k)
```

#### b) Generación de datos

```python
import math

pos= []
vel=[]
acel=[]
fuerza=[]
tiempo=[]

#Antes de abrir el paracaídas
p=altura_inicial
v=rapidez_inicial
a=-gravedad
t=0
dift=1

while(t < tiempo_parac):
    v=rapidez_inicial+a*t
    p=altura_inicial + (rapidez_inicial+v)*tiempo_parac/2
    f=masa_parac*a
    fuerza.append(f)
    pos.append(p)
    vel.append(v)
    acel.append(a)
    tiempo.append(t)
    t = t + dift

print("Antes de abrir el paracaídas:")
print("Tiempo: {:.2f},  Posición: {:.2f}, Velocidad: {:.2f}, Aceleración: {:.2f}".format(tiempo[-1],p,v,a))


#Al abrir el paracaídas

t = 0
dift = 0.01
p = altura_inicial + (rapidez_inicial + v) * tiempo_parac / 2 
v = rapidez_inicial - gravedad * tiempo_parac
a = (masa_parac * -gravedad + k * pow(v, 2)) / masa_parac
f=masa_parac*a

tiempo.append(t + tiempo_parac)
pos.append(p)
vel.append(v)
acel.append(a)
fuerza.append(f)

print("Al abrir el paracaídas:")
while ( a > 0):
    v_ini = v
    t = t + dift      
    v = v + a * t
    a = (masa_parac * -gravedad + k * pow(v, 2)) / masa_parac
    f = masa_parac * a
    p = p + ((v_ini + v) * dift / 2)
    f=masa_parac*a
    fuerza.append(f)
    tiempo.append(t + tiempo_parac)
    vel.append(v)
    acel.append(a)
    pos.append(p)
    
  
    print("Tiempo: {:.2f}, Posición: {:.2f}, Velocidad: {:.2f}, Aceleración: {:.2f}".format(t, p,v,a))
    

#Después de abrir el paracaídas

print("Después de abrir el paracaídas:")
equilibrio = tiempo_parac + t
pos_equi = p
t = equilibrio
dift = 1
a = 0
while (p > 0):    
    p =pos_equi + (t - equilibrio) * v
    f=masa_parac*a
    fuerza.append(f)
    tiempo.append(t + tiempo_parac)
    vel.append(v)
    acel.append(a)
    pos.append(p)
    t = t + dift
    print("Tiempo: {:.2f}, Posición: {:.2f}, Velocidad: {:.2f}, Aceleración: {:.2f}".format(t,p, v, a ))

    
```

#### c) Gráficos

##### Posición vs Tiempo


```python
import matplotlib.pyplot as plt
plt.plot(tiempo, pos)
plt.show()
```

##### Velocidad vs Tiempo

```python
import matplotlib.pyplot as plt
plt.plot(tiempo, vel)
plt.show()
```

##### Aceleración vs Tiempo

```python
import matplotlib.pyplot as plt
plt.plot(tiempo, acel)
plt.show()
```

##### Fuerza de Arrastre vs Tiempo

```python
import matplotlib.pyplot as plt
plt.plot(tiempo, fuerza)
plt.show()
```

# 2. Movimiento de un Proyectil

#### a) Parámetros

```python
import random
pos_ini_x=0
pos_ini_y=0
rapidez_inicial =50
radio=0.0366
area =math.pi*(radio**2)
resistencia=0.50
angulo=random.uniform(20.00,70.01)
densidad_aire=1.20
gravedad = 9.81
masa=0.145
k = (resistencia * densidad_aire * area) / 2


print("Posición inicial X: ", pos_ini_x)
print("Posición inicial Y: ", pos_ini_y)
print("Rapidez inicial: ", rapidez_inicial)
print("Área del proyectil:", area)
print("Coeficiente de resistencia: ", resistencia)
print("Ángulo de lanzamiento: ", angulo) 
print("Densidad del aire: ", densidad_aire)
print("Masa: ", masa)
print("K: ", k)
```

#### b)Generación de datos

```python
import math

posx= []
posy= []
vel= []
velx= []
vely= []
acex= []
acey= []
fuerzax= []
fuerzay= []
tiempo= []


theta = (angulo * math.pi) / 180
p_x = 0
p_y = 0
v = rapidez_inicial
v_x = v * math.cos(theta)
v_y = v * math.sin(theta)
a_x = 0
a_y = -gravedad
f_x = masa * a_x
f_y = masa * a_y
t = 0
dift = 0.01

posx.append(p_x)
posy.append(p_y)
vel.append(v)
velx.append(v_x)
vely.append(v_y)
acex.append(a_x)
acey.append(a_y)
fuerzax.append(f_x)
fuerzay.append(f_y)
tiempo.append(t)

while ( p_y >= 0):
    
    f_x = -k * v * v_x
    f_y = -k * v * v_y
    a_x = f_x / masa
    a_y = (masa * -gravedad + f_y) / masa
    v_x = v_x + a_x * dift
    v_y = v_y + a_y * dift
    p_x = p_x + ((velx[-1] + v_x) * dift / 2)
    p_y = p_y + ((vely[-1] + v_y) * dift / 2)
    v = math.sqrt(v_x * v_x + v_y * v_y)
    posx.append(p_x)
    posy.append(p_y)
    vel.append(v)
    velx.append(v_x)
    vely.append(v_y)
    acex.append(a_x)
    acey.append(a_y)
    fuerzax.append(f_x)
    fuerzay.append(f_y)
    tiempo.append(t)
    t = t + dift
    print("Tiempo:{:.2f} PosX:{:.2f} PosY:{:.2f} Velocidad:{:.2f} VelocidadX:{:.2f} VelocidadY:{:.2f} AceleraciónX:{:.2f} AceleraciónY:{:.2f}".format(t,p_x,p_y,v,v_x,v_y,a_x,a_y))
    
```

#### C)Gráficos

##### Posición X vs Posición Y

```python
import matplotlib.pyplot as plt
plt.plot(posx, posy)
plt.show()
```

##### Fuerza de Arrastre vs Tiempo

```python
import matplotlib.pyplot as plt
plt.plot(tiempo, fuerzax)
plt.plot(tiempo, fuerzay)
plt.show()
```
